### genres, id, keywords, overview, title_x, cast, crew

```python
df_id.columns
```

```python
df=df_id[["genres", "id", "keywords", "overview", "title_x", "cast", "crew"]]
```

```python
df.head(5)
```

```python
df.shape
```

```python
df=df[df["overview"].notna()]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
def convert(text):
    text=ast.literal_eval(text)
    ans=[]
    for i in text:
        ans.append(i["name"])
    return ans
```

```python
df["genres"]=df["genres"].apply(convert)
```

```python
df["keywords"]=df["keywords"].apply(convert)
```

```python
def convert3(text):
    text=ast.literal_eval(text)
    ans=[]
    cn=0
    for i in text:
        cn+=1
        ans.append(i["name"])
        if cn==3:
            break
    return ans
```

```python
df["cast"]=df["cast"].apply(convert3)
```

```python
l=(df["crew"][0])
```

```python
print(l)
```

```python
def fetch_director(text):
    text=ast.literal_eval(text)
    ans=[]
    
    for i in text:
        if i["job"]=="Director":
            ans.append(i["name"])
        
    return ans
```

```python
df["crew"]=df["crew"].apply(fetch_director)
```

```python
df.head(5)
```

```python
df["overview"]=df["overview"].apply(lambda x: x.split())
```

```python
df.head(5)
```

```python

df["genres"]=df["genres"].apply(lambda x: [i.replace(" ","") for i in x])
df["keywords"]=df["keywords"].apply(lambda x: [i.replace(" ","") for i in x])
df["overview"]=df["overview"].apply(lambda x: [i.replace(" ","") for i in x])
df["cast"]=df["cast"].apply(lambda x: [i.replace(" ","") for i in x])
df["crew"]=df["crew"].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
df.head(5)
```

```python
df["tag"]=df["genres"] + df["keywords"] + df["overview"] + df["cast"] + df["crew"]
```

```python
df.head(5)
```

```python
df=df[["id","title_x", "tag"]]
```

```python
df.head(5)
```

```python
df["tag"]=df["tag"].apply(lambda x: " ".join(x))
```

```python
df.head()
```

### Text preprocessing

```python
df["tag"]=df["tag"].apply(lambda x: x.lower())
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

```

```python
df['tag'].apply(lambda x: stemmer.stem(x))
```

```python
df.shape
```

```python
from  sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words="english", max_features=500)
```

```python
vector=cv.fit_transform(df["tag"]).toarray()
```

```python
#cv.get_feature_names_out()
```

## Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(vector)
```

```python
similarity_matrix.shape
```

```python
def recomend(Movie):
    index=df[df["title_x"]==Movie].index[0]
    movie_list=sorted(enumerate(similarity_matrix[index]), key=lambda x: x[1],reverse=True)[1:6]
    for i in movie_list:
        print(df["id"][i[0]])
        print(df["title_x"][i[0]])
```

```python
recomend("Spectre")
```
